```python
#import
import pygame
import time

#各種變數
WIN_WIDTH = 1024
WIN_HEIGHT = 600
BTN_WIDTH = 80
BTN_HEIGHT = 80
HP_WIDTH = 40
HP_HEIGHT = 40
FPS = 30

# color (RGB)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# initialization
pygame.init()

# create window surface
win = pygame.display.set_mode((WIN_WIDTH, WIN_HEIGHT))

# set window caption
pygame.display.set_caption("My first game")

# clock
clock = pygame.time.Clock()

# load image (background, enemy, buttons)
#匯入圖片
Map_image= pygame.image.load("Map.png")
enemy_image= pygame.image.load("enemy.png")
continue_image= pygame.image.load("continue.png")
hp_gray_image= pygame.image.load("hp_gray.png")
hp_image= pygame.image.load("hp.png")
muse_image= pygame.image.load("muse.png")
pause_image= pygame.image.load("pause.png")
sound_image= pygame.image.load("sound.png")

background_image = pygame.transform.scale(pygame.image.load("images/Map.png"), (WIN_WIDTH, WIN_HEIGHT))
#更改圖片尺寸
enemy_image= pygame.transform.scale(enemy_image,(60,60))
hp_gray_image= pygame.transform.scale(hp_gray_image,(HP_WIDTH,HP_HEIGHT))
hp_image= pygame.transform.scale(hp_image,(HP_WIDTH,HP_HEIGHT))
#muse_image= pygame.transform.scale(muse_image,())
#pause_image= pygame.transform.scale(pause_image,())
#sound_image= pygame.transform.scale(sound_image,())

class Game:
    def __init__(self):
        # window
        # ...(to be done)

        # hp
        self.hp = 7
        self.max_hp = 10
        pass

    def game_run(self):
        # game loop
        run = True
        while run:#迴圈
            clock.tick(FPS)
            # event loop#迴圈
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    run = False
                               
            # draw background
            #背景
            win.blit(Map_image, (0,0))
            
            # draw enemy and health bar
            #畫可愛小兵跟血條
            win.blit(enemy_image, (30,250))
            pygame.draw.rect(win,RED,[40,240,40,5])
            # draw menu (and buttons)
            #場景布置
            pygame.draw.rect(win,BLACK,[0,0,1024,80])
            win.blit(pause_image, (952,5))
            win.blit(continue_image, (892,5))
            win.blit(sound_image, (822,5))
            win.blit(muse_image, (755,5))
            win.blit(hp_image, (430,10))
            win.blit(hp_image, (465,10))
            win.blit(hp_image, (500,10))
            win.blit(hp_image, (535,10))
            win.blit(hp_image, (570,10))
            win.blit(hp_image, (430,40))
            win.blit(hp_image, (465,40))
            win.blit(hp_gray_image, (500,40))
            win.blit(hp_gray_image, (535,40))
            win.blit(hp_gray_image, (570,40))
            pygame.draw.rect(win,BLACK,[0,575,50,80])
            


            # draw time
            #加上計時器
            game_clock=pygame.time.get_ticks()
            #定義時間
            sec_clock=game_clock//1000
            sec_clock1=sec_clock%10
            sec_clock10=(sec_clock%60)//10
            min_clock=sec_clock//60
            min_clock1=min_clock%10
            min_clock10=(min_clock%60)//10
            #將(十與個)位數的分與秒透過運算定義出來
            font=pygame.font.SysFont("simhei",25)
            clock1=font.render(str(sec_clock1),BLACK,WHITE)
            clock2=font.render(str(sec_clock10),BLACK,WHITE)
            clock3=font.render(':',BLACK,WHITE)
            clock4=font.render(str(min_clock1),BLACK,WHITE)
            clock5=font.render(str(min_clock10),BLACK,WHITE)
            #貼上時鐘
            win.blit(clock1, (35,580))
            win.blit(clock2, (25,580))
            win.blit(clock3, (19,578))
            win.blit(clock4, (11,580))
            win.blit(clock5, (1,580))
            pygame.time.Clock()
            pygame.display.update()
if __name__ == "__main__":
    covid_game = Game()
    covid_game.game_run()
pygame.quit()#跳出
```
